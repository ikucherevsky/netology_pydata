```python
# Кусок кода из лекции для отбрасывания сильно коррелирующих признаков
# создаем матрицу корреляций
corr_matrix = balanced.corr().abs()

# Выбираем верхний треугольник матрицы
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))

# Ищем индексы колонок с корреляцией большей чем 0.9
to_drop = [column for column in upper.columns if any(upper[column] > 0.9)]

print(to_drop)
```

```python
# Заготовка из данных со всеми изначальными колонками
X_complete = balanced.iloc[:,:30].to_numpy() 
Y_complete = balanced.iloc[:,30:].to_numpy() 
```

```python
# удаляем сильно коррелирующие колонки методом drop()
balanced = balanced.drop(balanced[to_drop], axis=1)
balanced.head()
```

```python
X_balanced = balanced.iloc[:,:24].to_numpy() 
Y_balanced = balanced.iloc[:,24:].to_numpy() 
```

```python
# Здесь делим данные на обучающую выборку и тестовую , test_size = 0.2 задает долю тестовой выборки = 20%
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_balanced,
                                                 Y_balanced,
                                                 test_size = 0.2,
                                                 random_state = 3)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score 
model = LogisticRegression()
model.fit(X_train, y_train)
predicted_classes = model.predict(X_test)
accuracy = accuracy_score(y_test.flatten(),predicted_classes)
print('Точность предсказания модели: ', accuracy)
parameters = model.coef_
```
