```python
from scipy import stats
```

## Проверим есть ли различия в мощности японских автомобилей (в основном малолитражные, предполагается что сильных различий не будет)

```python
len(data[data['make']=='toyota'])
```

```python
pd.unique(data['make']).tolist()
```

```python
len(data[data['make']=='honda'])
```

```python
stats.ttest_ind(data[data['make']=='toyota']['horsepower'].sample(n=13),
                data[data['make']=='honda']['horsepower'].sample(n=13))
```

## p-value >> 0.05 , следовательно мощность этих японских автомобилей имеет значимую статистическую зависимость и в среднем нет различий в мощности Тойоты и Хонды

## Для Порше иная ситуация, в среднем мощности разные, что подтверждается тестом
Выборка, конечно должна быть побольше.

```python
len(data[data['make']=='porsche'])
```

```python
stats.ttest_ind(data[data['make']=='toyota']['horsepower'].sample(n=5),
                data[data['make']=='porsche']['horsepower'].sample(n=5))
```
