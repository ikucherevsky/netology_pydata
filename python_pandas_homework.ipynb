## 2

```python
power_data = pd.read_csv('power.csv')
Latvia_4_12_21 = power_data[
                           (power_data['country']=='Latvia')
                           & ((power_data['category'] == 4) | (power_data['category'] == 12) | (power_data['category'] == 21))
                           & (power_data['year'] >= 2005) & (power_data['year'] <= 2009)
                           & (power_data['quantity'] > 0)
                           ]
print(sum(Latvia_4_12_21['quantity']))
Latvia_4_12_21.head()
```

```python
Estonia_4_12_21 = power_data[
                           (power_data['country']=='Estonia')
                           & ((power_data['category'] == 4) | (power_data['category'] == 12) | (power_data['category'] == 21))
                           & (power_data['year'] >= 2005) & (power_data['year'] <= 2009)
                           & (power_data['quantity'] > 0)
                           ]
print(sum(Estonia_4_12_21['quantity']))
Estonia_4_12_21.head()
```

```python
Lithuania_4_12_21 = power_data[
                           (power_data['country']=='Lithuania')
                           & ((power_data['category'] == 4) | (power_data['category'] == 12) | (power_data['category'] == 21))
                           & (power_data['year'] >= 2005) & (power_data['year'] <= 2009)
                           & (power_data['quantity'] > 0)
                           ]
print(sum(Lithuania_4_12_21['quantity']))
Lithuania_4_12_21.head()
```

```python
baltic_summary = sum(Lithuania_4_12_21['quantity']) + sum(Latvia_4_12_21['quantity']) + sum(Estonia_4_12_21['quantity'])
print(f"суммарное потребление стран Балтии {baltic_summary}")
```
