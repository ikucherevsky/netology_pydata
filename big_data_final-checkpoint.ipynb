## 3. Оставьте в датафрейме ratings только те фильмы, который оценил данный пользователь 


```python
number_one_user = ratings[ratings['user_id'] == 405]
number_one_user.head()
```

```python
movie_count = ratings.groupby('item_id').count().reset_index()
movie_count.drop(columns=['rating','timestamp'],inplace=True)
movie_count.columns=['item_id','rating_count']
movie_count.head()
```

```python
movie_avg = ratings.groupby('item_id').mean().reset_index()
movie_avg.drop(columns=['user_id','timestamp'],inplace=True)
movie_avg.columns=['item_id','avg_rating']
movie_avg.head()
```
