Напишите код для моделирования 1000 подбрасываний игрального кубика. Начертите диаграму histogram. Почему график изменился? Как применить центральную предельную теорему?

```python
X1 = np.zeros((10000, 1))
for i in range(10000):
    X1[i,0] = rd.randint(1,6)
plt.hist(X1,6)
plt.show()
```

## Распределение стало равномерным, вследствие большого числа эксперимертов, видно, что выпадение каждой из сторон кубика равновероятно при достаточно большом количестве экспериментов

## Для применения ЦПТ найдем суммы 10 (например) последовательных исходов в сгенерированной последовательности X1

```python
X2 = np.zeros((1000, 1))
for i in range (1000):
    X2[i,0] = sum(X1[i*10:(i+1)*10])
plt.hist(X2,10)
plt.show()
```

## Как видно распределение этих сумм принимает вид нормального
